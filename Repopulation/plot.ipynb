```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
l_ntk = np.load('./results/loss_ntk_1000_0.npy')
a_ntk = np.load('./results/acc_ntk__1000_0.npy')
l_l_ntk = np.load('./results/loss_ntk_load_1000_0.npy')
a_l_ntk = np.load('./results/acc_ntk_load_1000_0.npy')
l_repop = np.load('./results/loss_vaetanh_1000_0.npy')
a_repop = np.load('./results/acc_vaetanh_1000_0.npy')
l_r = np.load('./results/loss_random_1000_0.npy')
a_r = np.load('./results/acc_random_1000_0.npy')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.ndimage.filters import gaussian_filter1d
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.plot(np.arange(50)+1,gaussian_filter1d(l_r[:,0],2),label = 'Random',alpha = 0.8)
plt.plot(np.arange(50)+1,gaussian_filter1d(l_repop[:,0],2),label = 'Generated',alpha = 0.8)
#plt.plot(l_repop[:,0],label = 'Repopulation')
plt.legend()
plt.xlabel('epoch',fontsize=15)
plt.ylabel('loss (train)',fontsize=15)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
sns.set_style("whitegrid")
plt.plot(np.arange(50)+1,gaussian_filter1d(l_r[:,1],2),label = 'Random',alpha = 0.8)
plt.plot(np.arange(50)+1,gaussian_filter1d(l_repop[:,1],2),label = 'Generated',alpha = 0.8)
#plt.plot(l_repop[:,0],label = 'Repopulation')
plt.legend()
plt.xlabel('epoch',fontsize=15)
plt.ylabel('loss (test)',fontsize=15)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
plt.plot(np.arange(50)+1,100*gaussian_filter1d(a_r,2),label = 'Random',alpha = 0.8)
plt.plot(np.arange(50)+1,100*gaussian_filter1d(a_repop,2),label = 'Generated',alpha = 0.8)
#plt.plot(a_repop,label = 'Repopulation')
plt.legend()
plt.xlabel('epoch',fontsize=15)
plt.ylabel(r'accuracy (%)',fontsize=15)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
sns.set_style("whitegrid")
plt.plot(np.arange(50)+1,gaussian_filter1d(l_ntk[:,0],2),label = 'Tangent Space (Random)',alpha = 0.8)
plt.plot(np.arange(50)+1,gaussian_filter1d(l_l_ntk[:,0],2),label = 'Tangent Space (Generated)',alpha = 0.8)
#plt.plot(l_repop[:,0],label = 'Repopulation')
plt.legend()
plt.xlabel('epoch',fontsize=15)
plt.ylabel('loss (train)',fontsize=15)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```
